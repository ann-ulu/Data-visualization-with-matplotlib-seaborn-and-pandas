```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
# This is a test. plot graph using matplotlib -pyplot
x = [2, 5, 7]
y = [9, 10, 12]
plt.plot(x, y)
plt.title("test plot")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
#This is a test. plot a double line graph
x = [2, 5, 7]
y = [9, 10, 12]
z = [4, 8, 15]
plt.plot(x, y)
plt.plot(x,z)
plt.title("test plot2")
plt.xlabel("x")
plt.ylabel("y" and "z")
plt.legend(["this is y", "this is z"])
plt.show()
```

```python
#This is a test for seaborn package. plot a double line graph
x = [2, 5, 7]
y = [9, 10, 12]
z = [4, 8, 15]
sns.set()
plt.plot(x, y)
plt.plot(x,z)
plt.title("test plot2")
plt.xlabel("x")
plt.ylabel("y" and "z")
plt.legend(["this is y", "this is z"])
plt.show()
```
